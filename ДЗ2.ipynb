```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
```

```python
random.shuffle(lst)
```

```python
data = pd.DataFrame({'whoAmI':lst})
data.head(20)
```

***Выше результат работы из ДЗ. ***






**1 способ**


```python
for i in lst:
  data.loc[data['whoAmI']!=i,i] = '0'
else:
  data.loc[data['whoAmI']==i,i] = '1'
```

```python
data.head()
```




**2 способ**




```python
data.loc[data['whoAmI']=='human','human'] = '1'
data.loc[data['whoAmI']=='robot','human'] = '0'
data.loc[data['whoAmI']=='human','robot'] = '0'
data.loc[data['whoAmI']=='robot','robot'] = '1'
```
